```python
import pandas as pd
import matplotlib. pyplot as plt
```

```python
df['general influence'].value_counts()
```

```python
value_counts = df['general influence'].value_counts()

ax = value_counts.plot.bar()

# Add count labels to the bars
for i, v in enumerate(value_counts):
    ax.text(i, v + 0.5, str(v), ha='center', va='bottom')

# Add the number of rows as a label on the plot
ax.text(0, max(value_counts) + 10, f'Total Rows: {len(df)}', ha='left', va='bottom')

plt.show()
```

```python
value_counts = df['general influence'].value_counts()
total_rows = len(df)

# Create a pie chart
fig, ax = plt.subplots()
ax.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%')

# Add count and percentage labels to the pie slices
for i, v in enumerate(value_counts):
    count_label = str(v)
    percent_label = f'{(v/total_rows)*100:.2f}%'
    ax.text(i, 0, count_label, ha='center', va='center')
    ax.text(i, -0.15, percent_label, ha='center', va='center')

# Add the number of rows as a title
ax.set_title(f'Total E: {total_rows}')

plt.show()
```
